```python
import numpy as np
import matplotlib.pyplot as plt

class Escena:
  def __init__(self, radio, color):
    self.radio = radio
    self.color = color
    self.punto_x = 0
    self.punto_y = 0

  def crear_escena(self):
    self.dibujar_escena()

  def cambiar_escena(self, n_radio=None, n_color=None):
    if n_radio is not None:
      self.radio = n_radio
    if n_color is not None:
      self.color = n_color
    self.dibujar_escena()

  def dibujar_escena(self):
    plt.clf()
    fig, ax = plt.subplots()
    ax.add_patch(plt.Circle((0, 0), self.radio, color=self.color, fill=False))
    plt.plot(self.punto_x, self.punto_y, 'ro')
    ax.set_aspect('equal')
    ax.set_xlim(-self.radio-1, self.radio+1)
    ax.set_ylim(-self.radio-1, self.radio+1)
    plt.grid(True)
    plt.show()

  def desplazar_punto(self, distancia, direccion):
    if direccion == 'arriba':
      self.punto_y += distancia
    elif direccion == 'abajo':
      self.punto_y -= distancia
    elif direccion == 'izquierda':
      self.punto_x -= distancia
    elif direccion == 'derecha':
      self.punto_x += distancia
    self.dibujar_escena()

  def simular(self, velocidad, direccion, pasos=10):
        for _ in range(pasos):
            self.desplazar_punto(velocidad, direccion)

escena = Escena(15, 'blue')
escena.crear_escena()

escena.cambiar_escena(15, 'green')
escena.desplazar_punto(1, 'arriba')
escena.simular(1, 'derecha')
```
